```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris=pd.read_csv('./Iris.csv')
```

```python
iris.drop('Id',axis=1,inplace=True)
```

```python
iris.head()
```

```python
plt.show()
```

```python
df = pd.DataFrame(iris)
plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'])
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
plt.bar(df['Species'], df['PetalLengthCm'])
plt.title('Bar Chart: Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
iris.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encode = LabelEncoder()
labels = label_encode.fit_transform(iris.Species)
iris["target"]=labels
```

```python
iris["Species"].value_counts()
```

```python
iris["target"].value_counts()
```

```python
import seaborn as sns  # Importing seaborn for better color palettes

sns.scatterplot(data=iris,x='SepalLengthCm',y='SepalWidthCm',hue='Species')
```

```python
sns.scatterplot(data=iris,x='SepalLengthCm',y='PetalLengthCm',hue='Species')
```

```python
sns.countplot(data=iris,x='Species')
```

```python
sns.displot(iris['SepalLengthCm'])
```

```python
correlation=iris.corr()
```

```python
fig = plt.figure(figsize=(5,5))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
plt.show()
```
